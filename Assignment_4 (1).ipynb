```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup

header = ['page #','location','url','district']
data = []

for i in range(1,3):
    quote_page = requests.get('https://en.tripadvisor.com.hk/Attractions-g294217-Activities-a_allAttractions.true-Hong_Kong.html'+str(i)) 
    print("\n***** Page " + str(i) +" in action *****")
    soup = BeautifulSoup(quote_page.content,'html.parser')
    for item in soup.find_all("section", attrs={"class":"fXyCS duCXC"}): 
         page_no = str(i)
         location = item.find("div",attrs={"class":"bUshh o csemS"}).get_text().strip() 
         url = item.find("div",attrs={"class":"fVbwn cdAAV cagLQ eZTON dofsx"}).a.get('href') 
         district = item.findAll("div", {"class" : "WlYyy diXIH fPixj"})
         district = district [1]
         data.append((page_no, location, url, district))

df = pd.DataFrame(data,
   columns = header
)
df.to_csv('HK_attractions.csv', sep='\t', encoding='utf-8')
df
```
